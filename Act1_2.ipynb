```python
df=pd.read_excel('Datos de Facturaci√≥n.xlsx')
df
```

```python
df.info()
```

```python
# Convertir la columna 'CVE_CLPV' a n√∫meros enteros con manejo de errores 'coerce'
df['CVE_CLPV'] = pd.to_numeric(df['CVE_CLPV'], errors='coerce')

# Aplicar los filtros despu√©s de la conversi√≥n
df = df.dropna(subset=['CVE_CLPV'])  # Eliminar filas con valores NaN
```

```python
# Aplicar los filtros
df1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
df2 = df[~df['CVE_VEND'].isin(['4', '5'])]
df3 = df[df['FECHA_ENT'] == '28/02/2022']
df4 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == 'E')]
df5 = df[['CVE_DOC', 'FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]
df6 = df[(df.index >= 7001) & (df.index <= 7099)]
df7 = df[df['CVE_VEND'].isin([1, 2])]
```

```python
#Convertir archivos filtrados a CSV.
df1.to_csv("df1.csv")
df2.to_csv("df2.csv")
df3.to_csv("df3.csv")
df4.to_csv("df4.csv")
df5.to_csv("df5.csv")
df6.to_csv("df6.csv")
df7.to_csv("df7.csv")
```
